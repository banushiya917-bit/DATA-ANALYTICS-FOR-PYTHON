```
import pandas as pd
import matplotlib.pyplot as plt

# Sample Dataset
data = {
    "Patient": ["A", "B", "C", "D", "E"],
    "Disease": ["Fever", "Diabetes", "Fever", "Cold", "Diabetes"],
    "Age": [20, 45, 30, 15, 50]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display Dataset
print("Hospital Patient Data")
print(df)

# Disease-wise Count
disease_count = df["Disease"].value_counts()

print("\nDisease Count:")
print(disease_count)

# Bar Chart
disease_count.plot(kind="bar")
plt.title("Patient Count by Disease")
plt.xlabel("Disease")
plt.ylabel("Number of Patients")
plt.show()
```
