```python
total_sales = df["Price"].sum()
print("Total Sales:", total_sales)
```

```python
product_sales = df.groupby("Category")["Price"].sum()
print(product_sales)
```

```python
product_sales.plot(kind='bar')
plt.title("product sales analysis")
plt.xlabel("products")
plt.ylabel("sales")
plt.show()
```
